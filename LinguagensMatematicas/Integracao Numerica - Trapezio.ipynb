```python
# Importando bibliotecas

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
def trapezio(f, a, b):
    
    h = b - a
    
    T = (h/2) * (f(a) + f(b))
    
    return T

def reta(f, x, a, b):
    
    part1 = (x - b)/(a - b)
    part2 = (x - a)/(b - a)
    
    return part1 * f(a) + part2 * f(b)

def f(x):
    return (1 + (np.cos(x))**2)**(1/2)
```

```python
# Delimitações da figura
fig = plt.figure(figsize=(10,5)) 

plt.ylim(0, 2) # delimitando os pontos do eixo y 
x = np.linspace(0, 48, 200) # determinando os pontos do eixo x
a = x[0]
b = x[-1]
funcao_x = f(x) 

# Visualizaçãço do problema
plt.plot(x, funcao_x, linestyle='-', label="Função f(x)", color = '#FF00FF', alpha=1) # plot da funcao

# Criar título central, título dos eixos e legendas
plt.title("Função f(x) no intervalo [0, 48]")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Plotar todo o gráfico com os comandos determinados anteriormente
plt.show()
```

```python
# CALCULANDO A INTEGRAL DA FUNÇÃO
X = sp.Symbol("x")
exp = 1 + (sp.cos(X)**2)**(1/2)
area_funcao = sp.integrate(exp, (X, 0, 48)).evalf()

# Delimitações da figura
fig = plt.figure(figsize=(10,5)) 

plt.ylim(0, 2) # delimitando os pontos do eixo y 
x = np.linspace(0, 48, 200) # determinando os pontos do eixo x
a = x[0]
b = x[-1]
funcao_x = f(x) 

# Visualizaçãço do problema
plt.plot(x, funcao_x, linestyle='-', label="Função f(x)", color = '#FF00FF', alpha=1) # plot da funcao
plt.plot(x, funcao_x, linestyle='--', label="Área de integração", color = 'black', alpha=0.9) # delimitação da área 
plt.plot([a, a], [0, funcao_x[1]], '--', color = "black", alpha=0.9) # delimitação da área 
plt.plot([b, b], [0, funcao_x[-1]], '--', color = "black", alpha=0.9) # delimitação da área 
plt.fill_between(x, funcao_x, label='Área de integração', color = "#FF00FF", alpha=0.2) # área sob curva 

# Criar título central, título dos eixos e legendas
plt.title("Função f(x) no intervalo [0, 48]")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Plotar todo o gráfico com os comandos determinados anteriormente
plt.show()

print(f"O valor da área dado pela integral da função é {area_funcao}")
#print(f"O valor da área dado pelo método do trapázio é {area_trapezio}")
```

```python
# CALCULANDO MÉTODO DO TRAPÉZIO
a = 0
b = 48
area_trapezio = trapezio(f, a, b)
reta_trapezio = reta(f, x, a, b)

# Delimitações da figura
fig = plt.figure(figsize=(10,5)) 

plt.ylim(0, 2) # delimitando os pontos do eixo y 
x = np.linspace(0, 48, 200) # determinando os pontos do eixo x
funcao_x = f(x) 
reta_trapezio = reta(f, x, a, b)

# Visualizaçãço do problema
plt.plot(x, funcao_x, linestyle='-', label="Função f(x)", color = '#FF00FF', alpha=1) # plot da funcao
plt.plot(x, reta_trapezio, linestyle='--', label="Delimitações do trapézio",color = 'black', alpha=0.9) # delimitação da área 
plt.plot([a, a], [0, funcao_x[1]], '--', color = "black", alpha=0.9) # delimitação da área 
plt.plot([b, b], [0, funcao_x[-1]], '--', color = "black", alpha=0.9) # delimitação da área 
plt.fill_between(x, reta_trapezio, label="Área do trapézio", color = "#FF00FF", alpha=0.2) # área sob curva 

# Criar título central, título dos eixos e legendas
plt.title("Função f(x) no intervalo [0, 48]")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Plotar todo o gráfico com os comandos determinados anteriormente
plt.show()

print(f"O valor da área dado pelo método do trapézio é {area_trapezio}")
```

```python
# CALCULANDO MÉTODO DO TRAPÉZIO
intervalos = [(0, 12), (12, 24), (24, 36), (36, 48)]

area_trapezio = []

# Delimitações da figura
fig = plt.figure(figsize=(10,5)) 

plt.ylim(0, 2) # delimitando os pontos do eixo y 
funcao_x = f(x) 

# Visualizaçãço do problema
plt.plot(x, funcao_x, linestyle='-', label="Função f(x)", color = '#FF00FF', alpha=1) # plot da funcao

cores = ["#FF00FF","#FF83FA", "#8B008B", "#BF3EFF"]

cor = 0
contador = 1

for a, b in intervalos:
    
    x2 = np.linspace(a, b, 50)
    area = trapezio(f, a, b)
    area_trapezio.append(area)
    
    reta_trapezio = reta(f, x2, a, b)
    
    plt.plot(x2, reta_trapezio, linestyle='--', color = 'black', alpha=0.9) # delimitação da área 
    plt.plot([a, a], [0, reta_trapezio[0]], '--', color = "black", alpha=0.9) # delimitação da área 
    plt.plot([b, b], [0, reta_trapezio[-1]], '--', color = "black", alpha=0.9) # delimitação da área 
    plt.fill_between(x2, reta_trapezio, label=f"Área do trapézio {contador}", color = cores[cor], alpha=0.4) # área sob curva 
    
    cor += 1
    contador += 1

# Criar título central, título dos eixos e legendas
plt.title("Função f(x) no intervalo [0, 48]")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Plotar todo o gráfico com os comandos determinados anteriormente
plt.show()

print(f"O valor da área dado pelo método do trapézio composto é {sum(area_trapezio)}")
```
